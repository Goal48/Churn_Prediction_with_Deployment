```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
import pickle
```

```python
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
# Drop unnecessary columns
data = data.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

```python
# Label encoding for categorical variables
label_encoder_gender = LabelEncoder()
data['Gender'] = label_encoder_gender.fit_transform(data['Gender'])
```

```python
# One-hot encoding for 'Geography'
one_hot_encoder_geo = OneHotEncoder(handle_unknown='ignore')
geo_encoded = one_hot_encoder_geo.fit_transform(data[['Geography']])
geo_encoded_df = pd.DataFrame(geo_encoded.toarray(), columns=one_hot_encoder_geo.get_feature_names_out(['Geography']))
```

```python
# Conbine the one-hot encoded columns with the original dataframe
data = pd.concat([data.drop(['Geography'], axis=1), geo_encoded_df], axis=1)
data.head()
```

```python
# Split into Independent and Dependent variables
X = data.drop('EstimatedSalary', axis=1)
y = data['EstimatedSalary']
```

```python
# Spliting the data into training and testing sets
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.2,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
# Scaled the Features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Save the encoder and Scaler
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)
with open('one_hot_encoder_geo.pkl','wb') as file:
    pickle.dump(one_hot_encoder_geo,file)
# Save the encoder and Scaler
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

ANN REGRESSION PROBLEM

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(64,activation='relu',input_shape=(X_train.shape[1],)))  ## 1st Hidden Layers
model.add(Dense(32,activation='relu')) ## 2nd Hidden Layers
model.add(Dense(1)) ## Output Layer
```

```python
model.summary()
```

```python
## Compile the model
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
loss = tf.keras.losses.MeanAbsoluteError()

model.compile(optimizer=opt,loss=loss,metrics=['mae'])
```

```python
# Set up Tensorboard 
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
regression_log_dir = "regressionlogs/fit" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback = TensorBoard(log_dir=regression_log_dir,histogram_freq=1)
```

```python
# Set up Early Stopping
early_stopping_callback = EarlyStopping(monitor='val_loss',patience=12,restore_best_weights=True)
```

```python
# Train the model   
history = model.fit(
    X_train,y_train,validation_data=(X_test,y_test),epochs=100,
    callbacks = [early_stopping_callback,tensorflow_callback]
)
```

```python
# Evaluate the model
test_loss, test_mae = model.evaluate(X_test,y_test)
print(f"Test Loss: {test_loss}")
print(f"Test MAE: {test_mae}")
```
